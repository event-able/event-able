## Parsing events

```python
from xml.etree import ElementTree
import datetime
import characteristic as ch
```

```python
def parse_events(filename):
    root = ElementTree.fromstring(open('../data/events_rss.xml').read())
    channel = root.getchildren()[0]
    return [Event.parse(n)
            for n in channel.getchildren()
            if n.tag == 'item']
```

```python
@ch.attributes(['guid', 'link', 'category', 'title', 'description',
                'region', 'venue', 'isfree', 'date', 'tags'])
class Event(object):
    @staticmethod
    def parse(node):
        return Event(
            guid=node.find('guid').text,
            link=node.find('link').text,
            category=node.find('category').text,
            title=node.find('title').text,
            description=node.find('description').text,
            region=node.find('{myEvents}marketingRegion').text,
            venue=Venue.parse(node.find('{myEvents}venue')),
            isfree=node.find('{myEvents}freeEntry').text,
            date=parse_date(node.find('{myEvents}eventDate').text),
            tags=[t.text for t in node.findall('{myEvents}tags')],
        )

    def is_historical(self):
        return self.date < datetime.date.today()

    def to_dict(self):
        d = self.__dict__.copy()
        d['venue'] = d['venue'].to_dict()
        d['date'] = str(d['date'])
        return d


@ch.attributes(['name', 'address', 'latitude', 'longitude', 'area', 'suburb', 'city',
                'state', 'postcode'])
class Venue(object):
    @classmethod
    def parse(cls, node):
        address, lat, lon = cls.parse_address(node.find('address'))
        return Venue(
            name=node.attrib['name'],
            address=address,
            latitude=lat,
            longitude=lon,
            area=node.find('area').text,
            suburb=node.find('suburb').text,
            city=node.find('city').text,
            state=node.find('state').text,
            postcode=node.find('postcode').text,
        )

    def to_dict(self):
        return self.__dict__.copy()
    
    @staticmethod
    def parse_address(node):
        lines = [node.find('address1').text,
                 node.find('address2').text,
                 node.find('address3').text]
        lines = filter(None, lines)
        lat = node.find('latitude').text
        lon = node.find('longitude').text
        return lines, lat, lon

def parse_date(d):
    return datetime.date(int(d[:4]), int(d[4:6]), int(d[6:8]))

events = parse_events('../data/events_rss.xml')
```

### What tags are used?

```python
tags = Counter()
for e in events:
    for t in e.tags:
        tags[t] += 1

tags.most_common(20)
```

```python
import networkx as nx
```

```python
from itertools import combinations
weights = Counter()
for e in events:
    for t1, t2 in combinations(sorted(e.tags), 2):
        weights[t1, t2] += 1

g = nx.Graph()
for (t1, t2), weight in weights.iteritems():
    g.add_node(t1)
    g.add_node(t2)
    g.add_edge(t1, t2, weight=weight)
```

```python
nx.draw_networkx(g)
```

```python
nx.write_gml(g, 'tags.gml')
```
